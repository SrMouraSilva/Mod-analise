Deixando mais f√°cil acessar as informa√ß√µes do tamanho do pedalboard

Adicionado:
 * **x_effect_relative**: Posi√ß√£o do efeito relativo ao mais extremo no eixo x
 * **y_effect_relative**: Posi√ß√£o do efeito relativo ao mais extremo no eixo y

```python
pedalboard_pluginss = pd.read_json(pedalboards['data'].to_json(), orient='index')['plugins']

#plugins = [plugin for pedalboard_plugin in pedalboard_pluginss for plugin in pedalboard_plugins]
plugins = []

elements = lambda elements: pd.DataFrame([plugin for plugin in pedalboard_pluginss[index]])

for index in pedalboard_pluginss.index:
    max_x = max(pd.DataFrame([plugin for plugin in pedalboard_pluginss[index]])['x'])
    max_y = max(pd.DataFrame([plugin for plugin in pedalboard_pluginss[index]])['y'])

    for plugin in pedalboard_pluginss[index]:
        plugin['x_effect_relative'] = plugin['x']/max_x
        plugin['y_effect_relative'] = plugin['y']/max_y
        plugins.append(plugin)

plugins = pd.DataFrame(plugins)

plugins
```

Plugins de √°udio mais utilizados

```python
plugins_by_uri = plugins.groupby(['uri'])

plugins_by_uri_sorted = reversed(sorted(list(plugins_by_uri), key=lambda plugins_a: len(plugins_a[1])))

pd.DataFrame([(uri, len(df)) for uri, df in plugins_by_uri_sorted], columns=['uri', 'count'])
```

## Distribui√ß√£o dos efeitos nos pedalboards

A distribui√ß√£o de todos os efeitos considerando os dois eixos pouco serve, pois os patches possuem tamanhos diferentes e os dados obtidos **n√£o possuem essa informa√ß√£o** (somente possuem o tamanho do screenshot, que costuma ser menor que o tamanho do pedalboard).

Entretanto, no eixo x √© bastante √∫til, pois _geralmente_ revela sua ordem (da esquerda para direita).

Utilidade: Podemos sugerir plugins de √°udio pelos espa√ßos livres

```python
import matplotlib.pyplot as plt
%matplotlib inline

x = plugins['x']
y = plugins['y']

plt.scatter(x, y, label='plugin', color='b', marker='o')
plt.title('Posi√ß√£o dos plugins de √°udio')
plt.show()
```

```python
import colorsys

total_plugins = len(plugins_by_uri)

listaHSV = [(x*1.0/total_plugins, 0.5, 0.5) for x in range(total_plugins)]
listaRGB = list(map(lambda x: colorsys.hsv_to_rgb(*x), listaHSV))

index = 0
for plugin_group in plugins_by_uri:
    uri = plugin_group[0]
    df = plugin_group[1]
    legend = uri.split('/')[-1]

    x, y = df['x'], df['y']

    plt.scatter(x, y, label=legend, color=listaRGB[index], marker='o')
    index += 1

#plt.legend()
plt.title('Posi√ß√£o dos plugins de √°udio por tipo')
plt.show()
```

```python
def plot_uri(uri):
    plugin = plugins[plugins.uri == uri]

    x = plugin['x']
    y = plugin['y']

    plt.scatter(x, y, label='plugin', color='b', marker='o')
    plt.title('Posi√ß√£o dos plugins de √°udio: ' + uri)
    plt.xlabel('Eixo x')
    plt.ylabel('Eixo y')
    plt.show()

plot_uri('http://moddevices.com/plugins/mod-devel/Gain2x2')
plot_uri('http://moddevices.com/plugins/sooperlooper')
```

```python
def plot_y_uri(uri):
    plugin = plugins[plugins.uri == uri]

    split = uri.split('/')
    title = split[-2] + '/' + split[-1]
    
    plt.title(title)
    plt.xlabel('Eixo x')
    plt.ylabel('Quantidade')
    plt.show(plt.hist(plugin['x_effect_relative']))

plot_y_uri('http://moddevices.com/plugins/mod-devel/Gain2x2')
plot_y_uri('http://moddevices.com/plugins/caps/CabinetIV')
plot_y_uri('http://moddevices.com/plugins/mod-devel/SwitchBox2')
```
