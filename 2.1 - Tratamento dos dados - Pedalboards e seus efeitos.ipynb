## API v1 - Pedalboards

```python
import scrapy
from scrapy.crawler import CrawlerProcess
from scrapy.utils.project import get_project_settings
from scrapy.utils.log import configure_logging

import json


class PedalboardsV1Spider(scrapy.Spider):
    name = "pedalboards"
    start_urls = [
        'https://pedalboards.moddevices.com/api/pedalboards?skip=0&take=10',
    ]
    current = 0
    step = 25

    def parse(self, response):
        data_json = json.loads(response.body_as_unicode())

        for pedalboard in data_json['items']:
            yield pedalboard

        self.current += 1

        if data_json['hasMore']:
            next_page = 'https://pedalboards.moddevices.com/api/pedalboards?skip={}&take={}'.format(
                self.current*self.step,
                self.step
            )

            next_page = response.urljoin(next_page)
            yield scrapy.Request(next_page, callback=self.parse)

process = CrawlerProcess({
    'USER_AGENT': 'Mozilla/4.0 (compatible; MSIE 7.0; Windows NT 5.1)',
    'FEED_FORMAT': 'json',
    'FEED_URI': 'mod-pedalboards.json',
})

process.crawl(PedalboardsV1Spider)
process.start() # the script will block here until the crawling is finished
```
